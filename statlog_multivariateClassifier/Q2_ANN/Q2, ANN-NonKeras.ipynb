```python
import csv
from sklearn.svm import SVC
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def read_shuttle_dataset(filename="./satimage/sat.trn"):
    with open(filename) as file:
        #reader = csv.reader(file)
       # for row in reader:
         #   x.append(list(map(int,row[:-1])))
         #   y.append(int(row[-1]))
        x = pd.read_csv(file, delimiter = ' ', header=None)
    return x,          
   # return x, y

#print(tf.version)
#lsat = np.loadtxt("./satimage/sat.trn", delimiter = ' ')
#print(lsat.shape)

x_dataframe =  pd.read_csv("./satimage/sat.trn", usecols = [*range(0,36)], delimiter = ' ', header=None)
y_dataframe =  pd.read_csv("./satimage/sat.trn", usecols = [36], delimiter = ' ', header=None)

#print(y_dataframe_)

#Removing class 6 due to its abscence in the dataset and validity doubts
#Changing any instance of class 7 into class 6, class 6 is now "very damp grey soil"
y_dataframe = y_dataframe.replace(7,6)

x = x_dataframe.values.tolist()
y = y_dataframe.stack().tolist()


 
print(y)

```

```python
def to_categorical(data):
    new_data = []
    #classes = len(set(data))
    classes = 6
    #print(classes)
   
    for row in data:
        new_row = [0] * classes
        #print(row)
        new_row[row-1] = 1
        new_data.append(new_row)
    return new_data

y = to_categorical(y)
#new_row[row-1] = 1 due to the fact that our classes start from 1 to 6 and any indexing starts from 0
#7 as an index is o
```

```python
# Create the model, 'None' denotes that the first dimension, corresponding to the batch size, can be of any size
x_ = tf.placeholder(tf.float32, [4435, 36])
y_ = tf.placeholder(tf.float32, [None, 6])

# Create first layer weights
layer_0_weights = tf.Variable(tf.random_normal([36, 20]))
layer_0_bias = tf.Variable(tf.random_normal([20]))
layer_0 = tf.nn.sigmoid(tf.add((tf.matmul(x_, layer_0_weights)), layer_0_bias))

# Create second layer weights
layer_1_weights = tf.Variable(tf.random_normal([20, 13]))
layer_1_bias = tf.Variable(tf.random_normal([13]))
layer_1 = tf.nn.sigmoid(tf.add(tf.matmul(layer_0, layer_1_weights), layer_1_bias))

# Create third layer weights
layer_2_weights = tf.Variable(tf.random_normal([13, 6]))
layer_2_bias = tf.Variable(tf.random_normal([6]))
layer_2 = tf.nn.sigmoid(tf.add(tf.matmul(layer_1, layer_2_weights), layer_2_bias))

# Define error function
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_, logits=layer_2))

# Define optimizer and its task (minimise error function)
optimizer = tf.train.GradientDescentOptimizer(learning_rate=.2).minimize(cost)
```

```python
#Training test data

N_EPOCHS = 150000

sess = tf.InteractiveSession()
tf.global_variables_initializer().run()

print('Training...')

errors = []

# Train
for i in range(N_EPOCHS):
    _, error = sess.run([optimizer,cost], feed_dict={x_: x, y_: y})
    errors.append(error)
        
plt.plot(errors)
plt.show()        
```

```python
# Test trained model
tf.InteractiveSession().close()
correct_prediction = tf.equal(tf.argmax(layer_2, 1), tf.argmax(y_, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
print('accuracy: {0:f}'.format(sess.run(accuracy, feed_dict={x_:x, y_:y})))
```
